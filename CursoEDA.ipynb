```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/train_titanic.csv'
```

```python
titanic_data_set=pd.read_csv(url,error_bad_lines=False)
```

```python
titanic_data_set.shape
titanic_data_set.info()
```

```python
#missing data 
total = titanic_data_set.isnull().sum().sort_values(ascending=False)
```

```python
#porcentaje de mujeres 
#percentage of women survived
women = titanic_data_set.loc[titanic_data_set.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)
#porcentajes de hombres 
men = titanic_data_set.loc[titanic_data_set.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)
```

```python
titanic_data_set['Survived'] = titanic_data_set['Survived'].map({0:"not_survived",
1:"survived"})  #conteo de los supervivientes 
```

```python
#Graficación 
fig, ax = plt.subplots(1, 2, figsize = (10, 8))
titanic_data_set["Sex"].value_counts().plot.bar(color = "skyblue", ax =
ax[0])
ax[0].set_title("Número de Pasajeros por sexo")
ax[0].set_ylabel("# pasajeros")
sns.countplot("Sex", hue = "Survived", data = titanic_data_set, ax = ax[1])
ax[1].set_title("Sexo: Sobrevivientes vs Muertes")
plt.show()
```

```python
#distribución de la edad 
sns.distplot(titanic_data_set['Age'].dropna())
```

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(titanic_data_set,vars = [ 'Fare','Age','Pclass'],
hue="Survived")
plt.show()
```

```python
titanic_data_set['Embarked'] = titanic_data_set['Embarked'].map({"S":1,
"C":2,"Q":2,"NaN":0})
Tcorrelation = titanic_data_set.corr(method='pearson')
Tcorrelation
```

```python
sns.heatmap(Tcorrelation,xticklabels=Tcorrelation.columns,
 yticklabels=Tcorrelation.columns)
```

```python
# load time series dataset
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
url='https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/df_power.csv'
```

```python
df_power_system= pd.read_csv(url,error_bad_lines=False);
```

```python
df_power_system.columns
```

```python
df_power_system.shape
```

```python
df_power_system.info()
```

```python
#convertir  object para  datetime 
df_power_system['Date'] = pd.to_datetime(df_power_system['Date'])
df_power_system = df_power_system.set_index('Date')
# Add columns with year, month, and weekday name
df_power_system['Year'] = df_power_system.index.year
df_power_system['Month'] = df_power_system.index.month
df_power_system['Weekday Name'] = df_power_system.index.day_name()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_power_system['Consumption'].plot(linewidth=0.5)
cols_to_plot = ['Consumption', 'Solar', 'Wind']
axes = df_power_system[cols_to_plot].plot(marker='.', alpha=0.5,
linestyle='None',figsize=(14, 6), subplots=True)
```

```python
ax = df_power_system.loc['2016', 'Consumption'].plot()
#ax.set_ylabel('Daily Consumption (GWh)');
```

```python
ax = df_power_system.loc['2016-12', 'Consumption'].plot(marker='o', linestyle='-')
#ax.set_ylabel('Consumo Diario (GWh)');
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 7), sharex=True)
for name, ax in zip(['Consumption', 'Solar', 'Wind'], axes):
    sns.boxplot(data=df_power_system, x='Month', y=name, ax=ax)
    ax.set_ylabel('GWh')
```
